```python
import pandas as pd
```

```python
bicimad = pd.read_csv('datasets/bicimad.csv')
bicimad
```

```python
# new data frame with split value columns
bicimad_new = bicimad["geometry_coordinates"].str.split(",", n = 1, expand = True)
  
# making separate first name column from new data frame
bicimad["longitud"]= bicimad_new[0]
  
# making separate last name column from new data frame
bicimad["latitud"]= bicimad_new[1]
  
# Dropping old Name columns
bicimad.drop(columns =["geometry_coordinates"], inplace = True)
bicimad_new
```

```python
bicimad['longitud'] = bicimad['longitud'].str.replace('[', '', regex=True)
bicimad['latitud'] = bicimad['latitud'].str.replace(']', '', regex=True)
bicimad
```

```python
residentes = pd.read_csv('datasets/residentes.csv', sep=";")
residentes
```

```python
residentes.info()
```

```python
residentes = residentes[residentes['NUM'].notna()]
```

```python
residentes.info()
```

```python
publicos_dis = pd.read_csv('datasets/publicos-dis.csv', sep=";")
publicos_dis.head()
```

```python
publicos_dis.info()
```

```python
aparcamientos = pd.concat ([residentes, publicos_dis])
aparcamientos
```

```python
aparcamientos.info()
```

```python
parks = aparcamientos.drop(aparcamientos.columns[[2, 3, 4, 5,6,7,8,11,13,14,15,16,22,23,26,27,28,29,30]], axis=1)
parks
```

```python
parks['NUM'] = parks['NUM'].replace(['24-B'],'24')
parks['NUM'] = parks['NUM'].replace(['10 C'],'10')
parks['NUM'] = parks['NUM'].replace(['74 B'],'74')
parks['NUM'] = parks['NUM'].replace(['6-B'],'6')
parks['NUM'] = parks['NUM'].replace(['10-B'],'10')
parks['NUM'] = parks['NUM'].replace(['25-B'],'25')
parks['NUM'] = parks['NUM'].replace(['3-B'],'3')
parks['NUM'] = parks['NUM'].replace(['14-B'],'14')
parks['NUM'] = parks['NUM'].replace(['64-B'],'64')
parks['NUM'] = parks['NUM'].replace(['61 - B'],'61')
parks['NUM'] = parks['NUM'].replace(['25 -B'],'25')
parks['NUM'] = parks['NUM'].replace(['6 - b'],'6')
parks['NUM'] = parks['NUM'].replace(['1-B'],'1')
parks['NUM'] = parks['NUM'].replace(['13-B'],'13')
parks['NUM'] = parks['NUM'].replace(['143-B'],'143')
parks['NUM'] = parks['NUM'].replace(['2-B'],'2')
parks['NUM'] = parks['NUM'].replace(['4-B'],'4')
parks['NUM'] = parks['NUM'].replace(['9-B'],'9')
parks['NUM'] = parks['NUM'].replace(['12-B'],'12')
parks['NUM'] = parks['NUM'].replace(['98-B'],'98')
parks['NUM'] = parks['NUM'].replace(['19-B'],'19')
```

```python
parks["NUM"] = pd.to_numeric(parks["NUM"])
```

```python
parks.info()
```

```python
parks['NUM'] = parks['NUM'].fillna(0)
```

```python
parks['NUM'] = parks['NUM'].astype(int)
```

```python
parks['NOMBRE-VIA'] = parks['NOMBRE-VIA'].replace(['GENERAL YAG&amp;Uuml;E'],'GENERAL YAGUE')
```

```python
parks["Place address"] = parks["CLASE-VIAL"] + " " + parks["NOMBRE-VIA"] + "," + " " + parks["NUM"].astype(str)
parks
```

```python
parks = parks.drop(parks.columns[[2, 3, 4,]], axis=1)
parks
```
